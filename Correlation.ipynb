```python

#Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set(color_codes=True)

import matplotlib.image as mpimg
from numpy.random import randn
from numpy.random import seed
from scipy.stats import pearsonr
from scipy.stats import spearmanr
```

```python
df=pd.read_csv('StudentsPerformance.csv')
df.head()
```

```python
df.describe()
```

```python
plt.figure(figsize = (12, 3))
heatmap= sns.heatmap(df.corr(),vmin=-1, vmax=1, annot = True,linecolor='black')
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':18}, pad=16);
plt.show()
```

```python
import matplotlib.cm as cm


plt.figure(figsize = (12, 3))
heatmap= sns.heatmap(df.corr(),vmin=-1, vmax=1, annot = True,cmap='viridis')
heatmap.set_title('Correlation Heatmap', fontdict={'fontsize':18}, pad=16);
plt.xticks(rotation=45)
plt.yticks(rotation = 0)
plt.show()
```

```python
data1= df["math score"]
data2= df["reading score"]
data3= df["writing score"]

plt.scatter(data2, data3)
```

```python
covariance= np.cov(data2, data3)
print(covariance)
```

```python
import seaborn
seaborn.set(style='whitegrid')
seaborn.scatterplot(x="reading score",
                    y="writing score",
                    data=df, hue="gender")
```

```python
sns.jointplot(data=df, x="reading score", y="writing score", hue="gender", kind="kde")
```

```python
sns.jointplot(data=df, x="reading score", y="writing score", hue="gender")
```

```python
corr,_=pearsonr(data2, data3)
print('Pearsons correlation: %.3f' % corr)
```

```python
df.corr(method="pearson")
```

```python
corr,_=spearmanr(data2, data3)
print('Spearmans correlation: %.3f' % corr)
```
